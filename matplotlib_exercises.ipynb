# Matplotlib Exercises

1. Use matplotlib to plot the following equation:

     y = x^2 − x + 2
- You'll need to write the code that generates the x and y points.
- Add an annotation for the point 0, 0, the origin.

```python
# import pyplot
import matplotlib.pyplot as plt

# generate data
x = range(-5,6)
y = [(x ** 2) - x + 2 for x in x]

# plot x and y
plt.plot(x,y)

# add limit to y-axis
plt.ylim(0,35)

# add origin annotation
plt.annotate('Origin', (0,0), (1,10), arrowprops = dict())

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add title
plt.title('$y = x^2 − x + 2$', fontsize = 16);
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- y=√x
- y=x^3
- y=2^x
- y=1/(x+1)


```python
# import math library
import math

# generate data
x = range(0, 11)
y = [math.sqrt(x) for x in x]

# plot x and y
plt.plot(x,y)

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add title
plt.title('$y = \sqrt{x}$', fontsize = 16);
```

```python
# generate data
x = range(-10, 11)
y = [x**3 for x in x]

# plot x and y
plt.plot(x,y)

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add title
plt.title('$y = x^3$', fontsize = 16);
```

```python
# generate data
x = range(0, 11)
y = [2**x for x in x]

# plot x and y
plt.plot(x,y)

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add title
plt.title('$y = 2^x$', fontsize = 16);
```

```python
# generate data
x = range(0, 11)
y = [1/(x+1) for x in x]

# plot x and y
plt.plot(x,y)

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# add title
plt.title(r'$y = \frac{1}{x+1}$', fontsize = 16);
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# generate data
x = range(0, 11)
y1 = [math.sqrt(x) for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [1/(x+1) for x in x]

# assign layout and position of subplot
plt.subplot(2,2,1)
# plot x and y1, assign color to curve
plt.plot(x,y1,
        color = 'b')
# add title with corresponding color
plt.title('$y = \sqrt{x}$',
         color = 'b')
# add gridlines
plt.grid('--')
# add axis limits
plt.xlim(0,10)
plt.ylim(0,10)
# assign layout and position of subplot
plt.subplot(2,2,2)
# plot x and y2, assign color to curve
plt.plot(x,y2,
        color = 'g')
# add title with corresponding color
plt.title('$y = x^3$',
         color = 'g')
# add gridlines
plt.grid('--')
# add axis limits
plt.xlim(0,10)
plt.ylim(0,10)
# assign layout and position of subplot
plt.subplot(2,2,3)
# plot x and y3, assign color to curve
plt.plot(x,y3,
        color = 'r')
# add title with corresponding color
plt.title('$y = 2^x$',
         color = 'r')
# add gridlines
plt.grid('--')
# add axis limits
plt.xlim(0,10)
plt.ylim(0,10)
# assign layout and position of subplot
plt.subplot(2,2,4)
# plot x and y4, assign color to curve
plt.plot(x,y4,
        color = 'y')
# add title with corresponding color
plt.title(r'$y = \frac{1}{x+1}$',
         color = 'y')
# add gridlines
plt.grid('--')
# add axis limits
plt.xlim(0,10)
plt.ylim(0,10)
# auto-arrange subplots
plt.tight_layout();
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# generate data
x = range(0, 11)
y1 = [math.sqrt(x) for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [1/(x+1) for x in x]

# plot curves with labels, colors, and markers
plt.plot(x,y1,
        color = 'b',
        marker = 'o',
        label = '$y = \sqrt{x}$')
plt.plot(x,y2,
        color = 'g',
        marker = 'v',
        label = '$y = x^3$')
plt.plot(x,y3,
        color = 'r',
        marker = 's',
        label = '$y = 2^x$')
plt.plot(x,y4,
        color = 'y',
        marker = 'p',
        label = r'$y = \frac{1}{x+1}$')
# add title
plt.title('Math Equations')
# add axis limits
plt.xlim(0,10)
plt.ylim(0,10)
# add legend
plt.legend();
```
