```python
# show only the audi manufactured vehicles with city mileage > 20
```

```python
audis = df[df.manufacturer == 'audi'] # shows only audis
audis[audis.cty > 20] # filters to only audis with cty > 20
```

```python
# alternatively:
df[(df.manufacturer == 'audi') & (df.cty > 20)]
# could also assign conditions to variables
# in that case parentheses wouldn't be necessary
# but they are absolutely necessary without variables
```

```python
# crosstabs vs pivot tables
    # crosstabs are summations of categoryA by categoryB
    # pivot tables are any kind of aggregation of valueC by categoryA and categoryB
```

```python
df.head()
```

```python
pd.crosstab(df.manufacturer, df['class'])
# gives summation
# number of audis that are compact is 15, number of chevys that are midsixe is 5, etc.
# single row of crosstab is each value of CategoryA
# single column of crosstab is each value of CategoryB
# each field/intersection is the count of rows that belong to both CatA and CatB
```

```python
# df.pivot_table(index = 'CatA', columns = 'CatB', values = 'numericColumnA')
# defaults to mean of the values argument/column we specified
# NaNs means there is no 2seater audis or VWs
# avg cty mileage for compact audis is 17.9
# avg cty mileage for dodge minivans is 15.8
pd.pivot_table(df, index='manufacturer', columns='class', values = 'cty')
```

```python
df[(df.manufacturer == 'audi') & (df['class'] == 'compact')].cty.mean()
# gives value of avg cty mileage for compact audis
# pivot table above gives that data and also data for other makes/classes
```
