#### 1. Use your `get_db_url` function to help you explore the data from the `chipotle` database.

#### 2. What is the total price for each order?

```python
chipotle['item_price'] = chipotle.item_price.str.replace('$','').astype(float)
chipotle.dtypes
```

```python
chipotle.pivot_table(aggfunc='sum', index='order_id', columns='item_name', values='item_price', margins=True)['All']
```

#### 3. What are the most popular 3 items?

```python
qty_ordered_per_item = chipotle.groupby('item_name').count()['quantity']
qty_ordered_per_item.sort_values(ascending=False).head(3)
```

#### 4. Which item has produced the most revenue?

```python
chipotle[['item_name', 'item_price']].groupby('item_name').agg('sum').idxmax()
```
