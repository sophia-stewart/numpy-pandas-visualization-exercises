#### 8. Using the `chipotle` dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import pandas as pd
import numpy as np

def get_db_url(db_name):
    from env import user, host, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
sql = 'SELECT * FROM orders'
url = get_db_url('chipotle')
chipotle = pd.read_sql(sql,url)
chipotle.head()
```

```python
# chipotle['item_price'] = chipotle.item_price.str.replace('$','').astype(float)
most_popular_items = chipotle.groupby('item_name').agg('sum').sort_values(by='quantity', ascending=False).drop(columns=['id', 'order_id']).rename(columns={'item_price':'revenue'}).head(4)
most_popular_items
```

```python
sns.barplot(data = most_popular_items, x = most_popular_items.index, y = 'revenue')
plt.title('Revenue From Most Popular Items')
plt.xticks(rotation=15);
```
